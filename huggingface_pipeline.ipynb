# 프로불편러와 연애하기

- 사용모델: `snunlp/KR-FinBert-SC`
- 감정분석 후 부정 감정 지수가 특정 수치를 넘어가면 이별
- 그 외 여러 원인들로 이별
- 추후 대화형 모델도 추가하여 더 정밀한 기능 추가 가능

```python
from transformers import pipeline
import time

model_name = 'alsgyu/sentiment-analysis-fine-tuned-model'

sentiment = pipeline(
    "sentiment-analysis",
    model=model_name
)
```

```python
sentiment.model.config.id2label
```

```python
print(sentiment("사랑"))
print(sentiment("좋아"))
print(sentiment("짜증나"))
print(sentiment("죽어"))
print(sentiment("행복해"))
print(sentiment("알았어"))
print(sentiment("그냥"))
```

```python
label_map = {
    "LABEL_0": "neutral",
    "LABEL_1": "negative",
    "LABEL_2": "positive"
}
```

```python
import random

def love_chat():
    print("-------------------------------------------------------")
    print("❤️조금 예민하긴 하지만 모든 것이 완벽한 상대와의 연애가 시작되었습니다❤️")
    print("이별통보를 당하지 않도록 답장에 신경써주세요.")
    print("------------------------------------------------------\n")

    love_name = input("애인의 이름을 알려주세요.")
    start_time = time.time()
    last_input_time = start_time

    positive_reply = [
        '너는 말을 정말 예쁘게 해 (｡˃ ᵕ ˂ )b',
        '오늘도 너 덕분에 기분이 좋아졌어 *.☆⸜(⑉˙ᗜ˙⑉)⸝🤍.* ',
        '너가 너무 좋아( ˘ ³˘)💕'
    ]

    common_reply = [
        '응 그래.',
        '그랬구나',
        '아 진짜?',
    ]


    while True:
        current_time = time.time()
        total_seconds = int(current_time - start_time)

        if total_seconds >= 100:
            print(f"\n{love_name}: 우리 연애한 지 벌써 100초나 됐어! 역시 우리는 천생연분인가봐❤️")
            break

        user_input = input("나: ")
        input_time = time.time()
        interval = int(input_time - last_input_time)

        # 답장 텀이 30초 이상일 경우 이별
        if interval > 30:
             print(f"\n{love_name}: ...뭐하는데 답장이 {interval}초나 걸려? 사랑이 식었나보네 또 나만 진심이었지. 우리 헤어져.")
             break

        else:
            last_input_time = input_time
            dandab = ['응', '그래']

            if user_input in dandab:
                print(f'{love_name}: 단답 뭐야? 나랑 연락하는 게 귀찮니? 또 나만 진심이었지. 우리 헤어져.')
                break

            elif user_input.count('ㅋ') >= 1:
                print(f'{love_name}: 도대체 어느 부분이 그렇게 웃긴거야? 우리 헤어져.')
                break

            elif user_input.count('?') >= 2:
                print(f'{love_name}: 물음표를 왜 그렇게 많이 써? 거슬리네. 우리 헤어져.')
                break

            elif user_input == 'ㅎ':
                print(f'{love_name}: 답장이 그게 뭐야? 성의가 너무 없잖아. 또 나만 진심이었지. 우리 헤어져.')
                break

            else:
                emotion = sentiment(user_input)[0]
                label = label_map[emotion['label']]
                score = emotion['score']

                if label == 'negative' and score > 0.45:
                    print(f'\n{love_name}: 말투가 왜 그래? 내가 싫어진 거야? 항상 나만 진심이고, 나만 좋아하는 것 같아. 차라리 나무랑 연애하는 게 나을 것 같아. 우리 헤어져.')
                    break

                elif label == 'positive':
                    print(f'{love_name}: {random.choice(positive_reply)}')

                else:
                    print(f'{love_name}: {random.choice(common_reply)}')


    end_time = time.time()
    total_seconds = int(end_time - start_time)

    print('\n--------------------------------------------------')
    print("[GAME OVER: 이별 통보]")
    print(f"{love_name}님이 당신을 차단했습니다.")
    print("--------------------------------------------------")
    print(f"총 연애 기간: {total_seconds // 60}분 {total_seconds % 60}초")

    if total_seconds < 60:
        print(f'{love_name}, 돌아와줘~!! ༼;´༎ຶ ۝༎ຶ`༽༼;´༎ຶ ۝༎ຶ`༽༼;´༎ຶ ۝༎ຶ`༽')

    print('=========================================================')
```
